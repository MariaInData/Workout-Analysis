```python
#Reading the file and converting it as a dict. 
import pandas as pd
import xmltodict
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
#Credits to flo.tausend on TowardsDataScience for some code and for a great idea!
```

```python
input_path = "/Users/moyma/Documents/export.xml" #'/Users/apple_health_export/export.xml'
with open(input_path, 'r') as xml_file:
    input_data = xmltodict.parse(xml_file.read())
```

```python
#Records list for general health data & imported as Pandas Data Frame
records_list = input_data['HealthData']['Record']
df_records = pd.DataFrame(records_list)
```

```python
#Workout list for workout data
workouts_list = input_data['HealthData']['Workout']
df_workouts = pd.DataFrame(workouts_list)
```

```python
#preview of workout dataframe
df_workouts
```

```python
#Convert selected columns to numeric so we can do calcuations
# convert just columns "..." and "..."
df_workouts[["@duration", "@totalDistance", "@totalEnergyBurned"]] = df_workouts[["@duration", "@totalDistance", "@totalEnergyBurned"]].apply(pd.to_numeric)
df_workouts.dtypes
#convert dates to actual datetime
format = '%Y-%m-%d %H:%M:%S %z'
df_workouts['@creationDate'] = pd.to_datetime(df_workouts['@creationDate'],format=format)
df_workouts['@startDate'] = pd.to_datetime(df_workouts['@startDate'], format=format)
df_workouts['@endDate'] = pd.to_datetime(df_workouts['@endDate'],format=format)
df_workouts.dtypes
```

```python
#Rename Activity Types 
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeCycling", "Cycling")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeRunning", "Running")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeYoga", "Yoga")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeOther", "Other")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeElliptical", "Elliptical")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeBasketball", "Basketball")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeFunctionalStrengthTraining", "Strength Training")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeTraditionalStrengthTraining", "Strength Training")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeCoreTraining", "Core Training")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeCooldown", "Cooldown")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeHighIntensityIntervalTraining", "HIIT")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeJumpRope", "Jump Rope")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeWalking", "Walking")

df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeStairClimbing", "Stair Climbing")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeHiking", "Hiking")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeSnowSports", "Snow Sports")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeGolf", "Golf")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeDownhillSkiing", "Skiing")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeRowing", "Rowing")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeCardioDance", "Dancing")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypePilates", "Pilates")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeTennis", "Tennis")
df_workouts["@workoutActivityType"]= df_workouts["@workoutActivityType"].replace("HKWorkoutActivityTypeSoccer", "Soccer")
```

```python
df_workouts.head(15)
```

```python
#drop unnecessary columns for dashboard
df_workouts.drop(df_workouts[['@sourceName','@durationUnit', "@sourceVersion", '@device', '@creationDate','@durationUnit',
                            '@totalDistanceUnit','@totalEnergyBurnedUnit', '@sourceVersion', '@device', '@creationDate',
                           'MetadataEntry', 'WorkoutEvent','WorkoutRoute']], axis=1, inplace=True)
```

```python
#add date column
df_workouts['Date'] = df_workouts['@startDate'].dt.date
#get day of week in string 
df_workouts['day'] = df_workouts['@startDate'].dt.strftime('%A')

#add length of workout
df_workouts['Length'] = (df_workouts['@endDate'] -  df_workouts['@startDate'])
df_workouts['Length']= df_workouts['Length'].dt.total_seconds()/60
df_workouts.round(2)
```

```python
#getting column names to rename them
df_workouts.columns
```

```python
#changing column names
df_workouts.rename(columns={"Activity Type":'@workoutActivityType' , '@duration': 'Duration', '@totalDistance': 'Distance',
       '@totalEnergyBurned': "Calories Burned", '@startDate': 'Start Date', '@endDate': 'End Date', 'day': 'Day of Week', 
                   'Length': 'Length','date': 'Date'}, inplace=True)
```

```python
#rounding all columns to 2 decimal points
df_workouts=df_workouts.round(2)
```

```python
df_workouts['Mile Pace'] = df_workouts['Distance']/ df_workouts['Duration']
```
